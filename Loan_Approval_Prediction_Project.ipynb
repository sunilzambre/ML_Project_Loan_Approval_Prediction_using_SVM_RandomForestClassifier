```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.fit_transform(x_train)
x_test=scalar.fit_transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
model1=KNeighborsClassifier()
model2=GaussianNB()
model3=SVC()
model4=DecisionTreeClassifier()
model5=RandomForestClassifier()
models=[model1,model2,model3,model4,model5]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay

for model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(model)
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print(accuracy_score(y_test, y_pred))
```

**Evaluation metrics of a classification model,specifically a RandomForestClassifier() model, on a test dataset with 185 samples.**

**(1) precision:** Precision is the ratio of true positive predictions to all positive predictions made by the model. 
    In this case, precision for class 0 (the first row) is 0.86, which means that out of all the samples predicted as class 0       by the model, 86% were actually class 0
    Precision for class 1 (the second row) is 0.77, which means that out of all the samples predicted as class 1 by the model,     77% were actually class 1.    <br>

**(2) recall:** Recall is the ratio of true positive predictions to all actual positive samples in the dataset. 
    In this case, recall for class 0 is 0.46, which means that out of all the actual class 0 samples in the dataset, the model     correctly predicted 46%. Recall for class 1 is 0.96, which means that out of all the actual class 1 samples in the dataset,     the model correctly predicted 96%.    <br>

**(3) f1-score :** F1-score is a harmonic mean of precision and recall, and is a way to balance between precision and recall. 
    It is calculated as 2 x (precision x recall) / (precision + recall). 
    In this case, the f1-score for class 0 is 0.60, and the f1-score for class 1 is 0.85.   <br>

**(4) support :** The number of samples in each class in the test dataset.                  <br>

**(5) accuracy:** The overall accuracy of the model on the test dataset, which is the ratio of correct predictions to
      total number of predictions. In this case, the accuracy is 0.78.                      <br>

**(6) macro avg :** The macro average of precision, recall, and f1-score across both classes.  <br>

**(7) weighted avg :** The weighted average of precision, recall, and f1-score across both classes, taking into account the number of samples in each class.   <br>

**(8) [[30 35], [5 115]]:** The confusion matrix of the model on the test dataset, which shows the number of true positive, false positive, false negative, and true negative predictions for each class. In this case, the model correctly predicted 30 samples as class 0 and 115 samples as class 1, but incorrectly predicted 35 samples as class 0 and 5 samples as class 1. <br>
